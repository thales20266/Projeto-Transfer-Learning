```python
import cv2
from google.colab.patches import cv2_imshow # Import cv2_imshow

# 1) Carregar a imagem colorida
img = cv2.imread("img1.jpg")

# Check if image loading was successful
if img is None:
    print("Error: Could not load image. Please check the file path.")
else:
    # 2) Converter para escala de cinza
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 3) Binarizar (thresholding)
    # Aqui, pixels >= 128 viram 255 (branco), abaixo viram 0 (preto)
    _, binary = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)

    # 4) Salvar os resultados
    cv2.imwrite("imagem_gray.jpg", gray)
    cv2.imwrite("imagem_bw.jpg", binary)

    # 5) Mostrar na tela usando cv2_imshow
    print("Imagem Original:")
    cv2_imshow(img)
    print("Cinza Imagem:")
    cv2_imshow(gray)
    print("Preto e Branco Imagem:")
    cv2_imshow(binary)

    # cv2.waitKey(0) # Not needed with cv2_imshow
    # cv2.destroyAllWindows() # Not needed with cv2_imshow
```
